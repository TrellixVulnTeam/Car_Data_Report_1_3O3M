```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data= pd.read_csv("dataV2.csv",sep=',')

print(data)

```

```python
missingtable= data.isnull()

missingMake=0
missingModel=0
missingYear=0
missingDoors=0

for i,j in missingtable.iterrows():
    if j["Make"]: missingMake+=1
    if j["Model"]: missingModel+=1
    if j["Year"]: missingYear+=1
    if j["Number of Doors"]: 
        missingDoors+=1
        print(data["Make"][i])
        print(data["Model"][i])
        print(data["Year"][i])
        print(data["Number of Doors"][i])
print(missingMake,missingModel,missingYear,missingDoors)
```

```python
labels= 'Missing', 'NotMissing'
sizes= [missingDoors,data.shape[0]-missingDoors]
colors= ['red','green']
plt.pie(sizes,labels=labels, colors=colors,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
makedict={}

for i in data["Make"]:
    if not (i in makedict): makedict[i]=1
    else: makedict[i]+=1

for i in makedict:
    
    print("-",i,makedict[i])
```
